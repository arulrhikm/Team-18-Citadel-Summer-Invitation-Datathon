```python
def plot_dataframe(df, title=''):
    fig, ax = plt.subplots(figsize=(15, 6))
    ax.plot(pd.to_datetime(df.index), df)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_minor_locator(mdates.MonthLocator())
    plt.legend(df.columns)
    plt.grid(True)
    if title != '':
        plt.title(title)
    plt.show()
```

```python
def get_linear_regression_residual(curve1, curve2, price_df):
    df1 = price_df[curve1]
    df2 = price_df[curve2]
    
    df1.replace([np.inf, -np.inf], np.nan, inplace=True)
    df2.replace([np.inf, -np.inf], np.nan, inplace=True)
    price_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    
    df1 = df1.dropna()
    df2 = df2.dropna()
    price_df = price_df.dropna()
    
    X = df1
    Y = df2
    X = sm.add_constant(X)
    
    model = sm.OLS(Y, X)
    results = model.fit()
    
    #parameters
    alpha = results.params.iloc[0]
    beta = results.params.iloc[1]
    
    #linear regression residuals
    residuals = Y - (alpha + beta*X[curve1])
    # residuals.plot()
    
    return residuals
```

```python
def std_linreg_residuals(residuals, window=100):
    rolling_mean = residuals.rolling(window=window).mean()
    rolling_std = residuals.rolling(window=window).std()
    rolling_zscore = (residuals - rolling_mean) / rolling_std
    
    return rolling_zscore
```

```python
def spread_trading(curve1, curve2, price_df, return_df, threshold, exposure):
    spread = get_linear_regression_residual(curve1, curve2, price_df)
    zscore = std_linreg_residuals(spread)
    mean_rev_df = pd.merge(return_df, zscore.rename('zscore'), left_index=True, right_index=True)
    
    def get_exposure_x(state, x, threshold, exposure):
        if state == 0:
            if x > threshold:
                return -exposure, 1
            elif x < -threshold:
                return exposure, -1
        elif state == 1 and x <= 0:
            return 0, 0
        elif state == -1 and x >= 0:
            return 0, 0
        return -state*exposure, state
    
    def get_exposure_y(state, x, threshold, exposure):
        if state == 0:
            if x > threshold:
                return exposure, 1
            elif x < -threshold:
                return -exposure, -1
        elif state == 1 and x <= 0:
            return 0, 0
        elif state == -1 and x >= 0:
            return 0, 0
        return state*exposure, state
    
    
    mean_rev_df[curve1 + 'exposure'] = 0
    mean_rev_df[curve2 + 'exposure'] = 0
    state_x = 0
    state_y = 0
    exposures_x = []
    exposures_y = []
    
    for x in mean_rev_df['zscore']:
        exposure_x, state_x = get_exposure_x(state_x, x, threshold, exposure)
        exposures_x.append(exposure_x)
        exposure_y, state_y = get_exposure_y(state_y, x, threshold, exposure)
        exposures_y.append(exposure_y)
        
    mean_rev_df[curve1 + 'exposure'] = exposures_x
    mean_rev_df[curve2 + 'exposure'] = exposures_y
    zscores_exposure = mean_rev_df[['zscore', curve1 + 'exposure', curve2 + 'exposure']]
    zscores_exposure.loc[:, curve1 + 'exposure'] = zscores_exposure.loc[:, curve1 + 'exposure'] / 100
    zscores_exposure.loc[:, curve2 + 'exposure'] = zscores_exposure.loc[:, curve2 + 'exposure'] / 100
    plot_dataframe(zscores_exposure, f'{(curve1, curve2)} Z-Score and Exposure')
    
    mean_rev_df[curve1 + 'pnl'] = mean_rev_df[curve1 + 'exposure'] * mean_rev_df[curve1]
    mean_rev_df[curve2 + 'pnl'] = mean_rev_df[curve2 + 'exposure'] * mean_rev_df[curve2]
    mean_rev_df[curve1 + 'cumulative_pnl'] = mean_rev_df[curve1 + 'pnl'].cumsum()
    mean_rev_df[curve2 + 'cumulative_pnl'] = mean_rev_df[curve2 + 'pnl'].cumsum()
    mean_rev_df[f'{(curve1,curve2)} total_pnl'] = mean_rev_df[curve1 + 'cumulative_pnl'] + mean_rev_df[curve2 + 'cumulative_pnl']
    
    pnl_df = mean_rev_df[[curve1 + 'cumulative_pnl', curve2 + 'cumulative_pnl', f'{(curve1,curve2)} total_pnl']]
    plot_dataframe(pnl_df, f'{(curve1, curve2)} Total P&L')
    
    return pnl_df
    
    
```

```python
def get_pair_df(curve1, curve2):
    price_data = data[[curve1, curve2]].dropna()
    return_data = price_data.pct_change().dropna()
    return price_data.iloc[1:], return_data
```

```python
all_pnl_df = []
for pair in pairs:
    curve1, curve2 = pair
    price_data, return_data = get_pair_df(curve1, curve2)
    pnl_df = spread_trading(curve1, curve2, price_data, return_data, 2, 100)
    all_pnl_df.append(pnl_df)
```

```python
pairs_pnl = all_pnl_df[0].iloc[:,-1 :]
for i in range(1, len(all_pnl_df)):
    pairs_pnl = pd.merge(pairs_pnl, all_pnl_df[i].iloc[:,-1 :], left_index=True, right_index=True)
    
plot_dataframe(pairs_pnl, 'Pairs Total P&L')
print(pairs_pnl.iloc[-1, :])
```

```python
total_pnl = pairs_pnl.sum(axis=1)
plot_dataframe(total_pnl.to_frame('Total P&L'), 'Total P&L')
print(total_pnl.iloc[-1])
```

```python
def calculate_sharpe_ratio(pnl, risk_free_rate=0):
    excess_returns = pnl - risk_free_rate
    sharpe_ratio = excess_returns.mean() / excess_returns.std()
    return sharpe_ratio
```

```python
calculate_sharpe_ratio(pairs_pnl)
```
