```python
%matplotlib inline
import pandas as pd
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

from statsmodels.tsa.vector_ar.vecm import *
import statsmodels.tsa.stattools as ts
import statsmodels.api as sm
import functools as ft
from statsmodels.tsa.vector_ar import *
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, Lasso, ElasticNet
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import metrics
from sklearn.model_selection import cross_val_score
import scipy.stats as stats
import seaborn as sns
import matplotlib.dates as mdates
from copulas.bivariate import Clayton, Gumbel, Frank
from copulas.multivariate import VineCopula
```

```python
def plot_dataframe(df, title):
    fig, ax = plt.subplots(figsize=(15, 6))
    ax.plot(pd.to_datetime(df.index), df)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_minor_locator(mdates.MonthLocator())
    plt.legend(df.columns)
    plt.grid(True)
    if title != '':
        plt.title(title)
    plt.show()
```

```python
general_drug_manufacturers = ['LLY', 'ABBV', 'AMGN', 'JNJ', 'PFE']
healthcare_plans = ['UNH', 'ELV', 'CI', 'CVS', 'HUM', 'CNC', 'MOH']
special_drug_manufacturers = ['ZTS', 'VTRS', 'NBIX', 'CTLT', 'ITCI', 'LNTH', 'ALKS', 'PRGO']
medical_devices = ['ABT', 'SYK', 'BSX', 'MDT', 'EW', 'DXCM', 'STE', 'ZBH', 'ALGN', 'PODD']
medical_supplies = ['ISRG', 'BDX', 'RMD', 'WST', 'HOLX', 'COO', 'BAX','TFX']
medical_care_facilities = ['HCA', 'UHS', 'THC', 'DVA', 'EHC', 'CHE', 'ENSG', 'ACHC']
medical_distribution = ['CAH', 'MCK', 'COR', 'OMI', 'HSIC', 'ZYXI', 'PDCO']

general_drug_manufacturers_df = yf.download(general_drug_manufacturers, start='2000-01-01', end='2024-03-01')['Adj Close']
healthcare_plans_df = yf.download(healthcare_plans, start='2000-01-01', end='2024-03-01')['Adj Close']
special_drug_manufacturers_df = yf.download(special_drug_manufacturers, start='2000-01-01', end='2024-03-01')['Adj Close']
medical_devices_df = yf.download(medical_devices, start='2000-01-01', end='2024-03-01')['Adj Close']
medical_supplies_df = yf.download(medical_supplies, start='2000-01-01', end='2024-03-01')['Adj Close']
medical_care_facilities_df = yf.download(medical_care_facilities, start='2000-01-01', end='2024-03-01')['Adj Close']
medical_distribution_df = yf.download(medical_distribution, start='2000-01-01', end='2024-03-01')['Adj Close']

plot_dataframe(general_drug_manufacturers_df, 'General Drug Manufacturers')
plot_dataframe(healthcare_plans_df, 'Healthcare Plans')
plot_dataframe(special_drug_manufacturers_df, 'Special Drug Manufacturers')
plot_dataframe(medical_devices_df, 'Medical Devices')
plot_dataframe(medical_supplies_df, 'Medical Supplies')
plot_dataframe(medical_care_facilities_df, 'Medical Care Facilities')
plot_dataframe(medical_distribution_df, 'Medical Distribution')
```

```python
general_drug_manufacturers_df.to_csv('processed_datasets/general_drug_manufacturers.csv')
healthcare_plans_df.to_csv('processed_datasets/healthcare_plans.csv')
special_drug_manufacturers_df.to_csv('processed_datasets/special_drug_manufacturers.csv')
medical_devices_df.to_csv('processed_datasets/medical_devices.csv')
medical_supplies_df.to_csv('processed_datasets/medical_supplies.csv')
medical_care_facilities_df.to_csv('processed_datasets/medical_care_facilities.csv')
medical_distribution_df.to_csv('processed_datasets/medical_distribution.csv')
```

```python
stock_desc = pd.read_csv('datasets/stock_descriptions.csv')
stock_desc
```

```python
sector_groups = stock_desc.groupby('Sector')['Symbol'].apply(list).to_dict()

for sector, symbols in sector_groups.items():
    print(f"{sector}: {symbols}")

sector_groups['TECHNOLOGY'] = ['ALG', 'SAP', 'AGCO', 'CNH', 'DE', 'CAT']
```

```python
for sector, symbols in sector_groups.items():
    sector_df = yf.download(symbols, start='2000-01-01', end='2024-03-01')['Adj Close']
    sector_df.to_csv(f'processed_datasets/food_{sector.lower()}.csv')
    plot_dataframe(sector_df, sector)
```

```python
stocks_etfs = pd.read_csv('datasets/all_stock_and_etfs.csv')
stocks_etfs
```

```python
merged_df = stocks_etfs.merge(stock_desc, left_on='Ticker_Symbol', right_on='Symbol', how='left')

# Create a dictionary to hold sector-specific DataFrames
sector_dfs = {}

# Group by sector
for sector, group in merged_df.groupby('Sector'):
    # Pivot to get stock tickers as columns with their Close prices
    pivot_df = group.pivot(index='Date-Time', columns='Ticker_Symbol', values='Close')
    sector_dfs[sector] = pivot_df

# sector_dfs contains DataFrames for each sector
for sector, df in sector_dfs.items():
    print(f"\nSector: {sector}")
    plot_dataframe(df, f"{sector} Sector")
```
